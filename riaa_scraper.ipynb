```python
import pandas as pd
import requests
```

```python
headers = {'User-agent': 'Custom User Agent'}
```

```python
url = 'https://www.riaa.com/gold-platinum/?tab_active=awards_by_album#search_section'
```

```python
response = requests.get(url, headers=headers)
```

```python
table = pd.read_html(response.text)
table
```

```python
no_details = table[0]
no_details.head()
```

```python
no_details.fillna(value='All', inplace=True)
no_details.head()
```

```python
no_details.columns = no_details.iloc[0]
no_details.head()
```
